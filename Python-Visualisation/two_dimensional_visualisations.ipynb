```python
# Computational packages
import numpy as np
import pandas as pd

# Visualisation packages
import matplotlib.pyplot as plt

#Hide Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Generating X and Y 2-D arrays
X,Y = np.meshgrid(np.linspace(-2, 2, 41), np.linspace(-1, 1, 21))

# Computating Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

plt.pcolor(Z) #Resulting image with pcolor
plt.show()
```

# Filled contour plots

```python
plt.subplot(2,2,1)
plt.contour(X, Y, Z)
plt.title('Default contour map')

plt.subplot(2,2,2)
plt.contour(X, Y, Z, 20)
plt.title('Contour with 20 contours')

plt.subplot(2,2,3)
plt.contourf(X, Y, Z)
plt.title('Default filled contour map')

plt.subplot(2,2,4)
plt.contourf(X, Y, Z, 20)
plt.title('Default filled contour with 20 contours')

# Display
plt.rcParams["figure.figsize"] = (15,7)
plt.tight_layout() #Improving spacing
plt.show()
```

# Styled filled contour plots

```python
plt.subplot(2,2,1)
plt.contourf(X,Y,Z,20, cmap='viridis')
plt.colorbar()
plt.title('Filled contour with cmap - Viridis')

plt.subplot(2,2,2)
plt.contourf(X,Y,Z,20, cmap='gray')
plt.colorbar()
plt.title('Filled contour with cmap - Gray')

plt.subplot(2,2,3)
plt.contourf(X,Y,Z,20, cmap='autumn')
plt.colorbar()
plt.title('Filled contour with cmap - Autumn')

plt.subplot(2,2,4)
plt.contourf(X,Y,Z,20, cmap='winter')
plt.colorbar()
plt.title('Filled contour with cmap - Winter')

# Display
plt.rcParams["figure.figsize"] = (15,7)
plt.tight_layout() #Improving spacing
plt.show()
```

# 2D Histogram of Auto MPG data using rectangular bins

```python
auto = pd.read_csv('data/mpg_auto_datasets.csv')
plt.hist2d(auto['hp'], auto['mpg'], bins=(20, 20), range=((40, 235), (8, 48))) #2D Histogram

plt.colorbar() # Adding color bar to the histogram

# Display
plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('2D Histogram plot of Auto MPG Data using rectangular bins')
plt.show()
```

# 2D Histogram of Auto MPG data using hexagonal bins

```python
plt.hexbin(auto['hp'], auto['mpg'], gridsize=(15, 12), extent=(40, 235, 8, 48))

plt.colorbar() # Adding color bar to the histogram

# Display
plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('2D Histogram plot of Auto MPG Data using hexagonal bins')
plt.show()
```

# 2D Image Visualisations

```python
img = plt.imread('data/astronaut_raw_image.jpg')
print("Shape of the image = %s" % str(img.shape))

plt.imshow(img) #Display the image

# Display style formatting
plt.axis('off')
plt.show()
```

# Pseudocolor plot from the astronaut image

```python
intensity = img.sum(axis=2) # Computation of the sum of RGB channels
print ("Shape of the intensity = %s" % str(intensity.shape))

plt.imshow(intensity, cmap='gray')
plt.colorbar()

# Display style formatting
plt.axis('off')
plt.show()
```

# Investigating various 'Extent' and 'Aspect' options in the image

```python
# First (Top left) subplot
plt.subplot(2,2,1)
plt.title('extent=(-1,1,-1,1),\naspect=0.5')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=0.5)

# Second (Top right) subplot
plt.subplot(2,2,2)
plt.title('extent=(-1,1,-1,1),\naspect=1')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=1)

# Third (Bottom left) subplot
plt.subplot(2,2,3)
plt.title('extent=(-1,1,-1,1),\naspect=2')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=2)

# Fourth (Bottom right) subplot
plt.subplot(2,2,4)
plt.title('extent=(-2,2,-1,1),\naspect=2')
plt.xticks([-2,-1,0,1,2])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-2,2,-1,1), aspect=2)

# Display style formatting
plt.rcParams["figure.figsize"] = (15,10)
plt.tight_layout()
plt.show()
```
