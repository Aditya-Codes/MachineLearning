```python
#Compuatational Packages
import numpy as np
import pandas as pd

#Visualisation Packages
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading MPG Auto Dataset
auto_dataset = pd.read_csv('data/mpg_auto_datasets.csv')
auto_dataset.info()
```

# Linear Regression between weight and hp

```python
sns.lmplot(x='weight', y='hp', data=auto_dataset)

# Display
#plt.rcParams["figure.figsize"] = (12,8)
plt.show()
```

```python
# Grouping Linear Regression by hue ('origin')
sns.lmplot(x='weight', y='hp', data=auto_dataset, hue='origin', palette='Set1')
plt.show()
```

```python
# Linear Regression between weight and hp grouped row-wise by 'origin'
sns.lmplot(x='weight', y='hp', data=auto_dataset, row='origin')
plt.show()
```

# Residuals of the Regression

```python
sns.residplot(x='hp', y='mpg', data=auto_dataset, color='green')

# Display
plt.rcParams["figure.figsize"] = (15,7)
plt.show()
```

# Higher Order Regressions

```python
# Scatter plot of 'weight' and 'mpg' using red circles
plt.scatter(auto_dataset['weight'], auto_dataset['mpg'], label='data', color='red', marker='o')

# Linear regression of order 1 between 'weight' and 'mpg' in blue
sns.regplot(x='weight', y='mpg', data=auto_dataset, label='order 1', color="blue", scatter=None)

# Kinear regression of order 2 between 'weight' and 'mpg' in green
sns.regplot(x='weight', y='mpg', data=auto_dataset, label='order 2', color="green", scatter=None, order=2)

# Display
plt.legend(loc='upper right')
plt.rcParams["figure.figsize"] = (15,7)
plt.show()
```
