```python
#Computational Packages
import numpy as np
import pandas as pd

#Visualisation Packages
import matplotlib.pyplot as plt
import seaborn as sns

#Hide Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading MPG Auto Dataset
auto_dataset = pd.read_csv('data/mpg_auto_datasets.csv')
auto_dataset.info()
```

# Strip plot of 'hp' grouped by 'cyl'

```python
plt.subplot(2,1,1)
sns.stripplot(x='cyl', y='hp', data=auto_dataset)

plt.subplot(2,1,2)
sns.stripplot(x='cyl', y='hp', data=auto_dataset, jitter=True, size=3) #Using jitter and smaller point size

#Display
plt.rcParams["figure.figsize"] = (15,7)
plt.show()
```

# Swarm Plot of 'hp' grouped by 'cyl'

```python
plt.subplot(2,1,1)
sns.swarmplot(x='cyl', y='hp', data=auto_dataset)

plt.subplot(2,1,2)
sns.swarmplot(x='hp', y='cyl', data=auto_dataset, hue='origin', orient='h') #Swarm plot

# Display
plt.rcParams["figure.figsize"] = (15,7)
plt.show()
```

# Violin Plot

```python
plt.subplot(2,1,1)
sns.violinplot(x='cyl', y='hp', data=auto_dataset)

plt.subplot(2,1,2)
sns.violinplot(x='cyl', y='hp', data=auto_dataset, inner=None, color='lightgray') #Added color of lightgray

sns.stripplot(x='cyl', y='hp', data=auto_dataset, jitter=True, size=1.5) #Overlaying strip plot on the violin plot

# Display
plt.rcParams["figure.figsize"] = (15,7)
plt.show()
```
